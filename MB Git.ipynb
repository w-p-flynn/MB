```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, output_file, show
from bokeh.io import output_file, show
from bokeh.layouts import column
from bokeh.plotting import figure
```

```python
df = pd.read_csv('D 9 20 2022 DATA.txt',sep = '\t', skiprows = 9,header=None)
df.head()
```

```python
# normalized data 
# instantiating the figure object
graph = figure(title="MILK BAR Data - Normalized")
w = 1750
h = 1000
graph = figure(width=w, height=h)
# define data
x = df.iloc[:,0]
y1 = df.iloc[:,1]/df.iloc[:,1].abs().max()
y2 = df.iloc[:,2]/df.iloc[:,2].max()
y3 = df.iloc[:,3]/df.iloc[:,3].max()
y4 = df.iloc[:,4]/df.iloc[:,4].max()
y5 = df.iloc[:,5]/df.iloc[:,5].abs().max()
y6 = df.iloc[:,6]/df.iloc[:,6].abs().max()

# plotting the line graphs
graph.line(x, y1, legend_label="Contact Force",line_color = 'black')
graph.line(x, y2, legend_label="Oral Pressure", line_color='green')
graph.line(x, y3, legend_label="Left Jaw", line_color='blue')
graph.line(x, y4, legend_label="Right Jaw", line_color='red')
graph.line(x, y5, legend_label="Swallow PVDF Sensor", line_color='orange')
graph.line(x, y6, legend_label="Channel 7", line_color='purple')

graph.legend.click_policy="hide"

# displaying the model
show(graph)
```
